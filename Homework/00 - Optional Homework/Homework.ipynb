```python
import pandas as pd
```

```python
df = pd.read_csv('interactive_data.csv')
df
```

- Nearly *two-thirds* of gun deaths are *suicides*.

```python
mort = df['Deaths'][0]
mort
```

```python
df.head()
```

```python
df_Intention = df.sort_values(by = 'Intent')
df_Intention 
```

```python
df_suicide = df_Intention[['Intent']]
df_suicide
```

```python
df_taux_suicide = df[(df['Intent'] == 'Suicide')]
df_taux_suicide
```

```python
taux_suicide = df_taux_suicide['Deaths'][108]
taux_suicide
```

```python
ratio = (taux_suicide)/(mort)
print('Le taux de suicide est de : ',ratio*100,'%')
```

- More than *85 percent* of suicide victims are *male*.

```python
df_taux_suicide_male = df[(df['Intent'] == 'Suicide') & (df['Gender'] == 'Male')]
df_taux_suicide_male
```

```python
taux_suicide_male = df_taux_suicide_male['Deaths'][180]

ratio_2 = (taux_suicide_male)/(taux_suicide)
```

```python
print('Le taux de suicide masculin est de : ',ratio_2*100,'%')
```

- Around *a third* of all gun deaths are *homicides*.

```python
df_taux_homicide = df[(df['Intent'] == 'Homicide')]
df_taux_homicide
```

```python
taux_homicide = df_taux_homicide['Deaths'][216]
taux_homicide
```

```python
ratio = (taux_homicide)/(mort)
print('Le taux d homicide est de : ',ratio*100,'%')
```

- Around *two-thirds* of homicide victims who are *males* in the *age-group of 15--34* are *black*.

```python
df_homicide_male_age_b = df[(df['Intent'] == 'Homicide')&(df['Gender'] == 'Male')&(df['Age'] == '15 - 34')&(df['Race'] == 'Black')]
taux_homicide_male_age_b = df_homicide_male_age_b['Deaths'][302]
df_homicide_male_age = df[(df['Intent'] == 'Homicide')&(df['Gender'] == 'Male')&(df['Age'] == '15 - 34')]
taux_homicide_male_age = df_homicide_male_age['Deaths'][300]
```

```python
ratio_three = (taux_homicide_male_age_b)/(taux_homicide_male_age)
print('Le taux de suicide des Hommes Black dans le groupe dâge 15--34 est de : ',ratio_three*100,'%')
```

- *Women* constitue only *15 percent* of the total *homicide* victims.

```python
taux_women = df_taux_homicide[(df_taux_homicide['Gender'] == 'Female')]

taux_women_ratio = taux_women['Deaths'][252]
                               
ratio_women =  taux_women_ratio / taux_homicide
```

```python
print('Le taux de suicide des Female est de : ',ratio_women*100,'%')
```
